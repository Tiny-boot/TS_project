### VAE

Tensorflow with Keras for
the VAE implementation. The VAE architecture is composed
of symmetric encoder and decoder architectures implementing
6 dense layers and 3 drop-out layers. The latent space dimension has been setup to 10% of the dimension of the original
problem with a minimum of 3 dimensions. The VAE has been
trained using the Adam optimizer.

https://www.kaggle.com/code/kmader/vae-to-detect-anomalies-on-digits

```python
class VAE(nn.Module):
    def __init__(self, input_dim, latent_dim=10):
        super(VAE, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU()
        )
        self.z_mean = nn.Linear(32, latent_dim)
        self.z_log_var = nn.Linear(32, latent_dim)
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 32),
            nn.ReLU(),
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, input_dim),
            nn.Sigmoid()
        )

    def encode(self, x):
        h = self.encoder(x)
        return self.z_mean(h), self.z_log_var(h)

    def reparameterize(self, mean, log_var):
        std = torch.exp(0.5 * log_var)
        eps = torch.randn_like(std)
        return mean + eps * std

    def decode(self, z):
        return self.decoder(z)

    def forward(self, x):
        mean, log_var = self.encode(x)
        z = self.reparameterize(mean, log_var)
        return self.decode(z), mean, log_var

def loss_function(recon_x, x, mean, log_var):
    recon_loss = nn.MSELoss()(recon_x, x)
    kld_loss = -0.5 * torch.sum(1 + log_var - mean.pow(2) - log_var.exp())
    return recon_loss + kld_loss


for i, dataset in enumerate(datasets, start=1):
    X_train = globals().get(f"X_train_n_{i}")
    X_test = globals().get(f"X_test_{i}")
    y_true = globals().get(f"y_true_{i}")


    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    X_train_scaled = np.array(X_train_scaled)
    X_test_scaled = np.array(X_test_scaled)

    X_train_tensor = torch.FloatTensor(X_train_scaled)
    X_test_tensor = torch.FloatTensor(X_test_scaled)

    input_dim = X_train_scaled.shape[1]
    latent_dim = 10
    model = VAE(input_dim=input_dim, latent_dim=latent_dim)
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    model.train()
    for epoch in range(50):
        optimizer.zero_grad()
        recon_batch, mean, log_var = model(X_train_tensor)
        loss = loss_function(recon_batch, X_train_tensor, mean, log_var)
        loss.backward()
        optimizer.step()

    model.eval()
    with torch.no_grad():
        recon_test, _, _ = model(X_test_tensor)
        reconstruction_error = torch.mean((recon_test - X_test_tensor) ** 2, dim=1).numpy()

    scores = (reconstruction_error - np.min(reconstruction_error)) / (np.max(reconstruction_error) - np.min(reconstruction_error))

    fpr, tpr, thresholds = roc_curve(y_true, scores)
    auc_score = auc(fpr, tpr)
    ap_score = average_precision_score(y_true, scores)

    results.loc[dataset, "VAE_AUC"] = auc_score
    results.loc[dataset, "VAE_AP"] = ap_score

auc_scores = []
ap_scores = []

for fault in faults:
    data = steel_plates_faults.data.original
    non_faults = [f for f in faults if f != fault]
    Xn = data[data[fault] == 1].drop(columns=faults)
    Xa = data[data[non_faults].sum(axis=1) == 0].drop(columns=faults)

    X_train_n, X_test_n = train_test_split(Xn, test_size=0.2, random_state=42)

    X_train_n = np.array(X_train_n)
    Xa = np.array(Xa)

    X_test = np.vstack((X_test_n, Xa))
    y_true = np.array([-1] * len(X_test_n) + [1] * len(Xa))

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_n)
    X_test_scaled = scaler.transform(X_test)

    X_train_scaled = np.array(X_train_scaled)
    X_test_scaled= np.array(X_test_scaled)

    X_train_tensor = torch.FloatTensor(X_train_scaled)
    X_test_tensor = torch.FloatTensor(X_test_scaled)


    input_dim = X_train_scaled.shape[1]
    latent_dim = 10
    model = VAE(input_dim=input_dim, latent_dim=latent_dim)
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    model.train()
    for epoch in range(50):
        optimizer.zero_grad()
        recon_batch, mean, log_var = model(X_train_tensor)
        loss = loss_function(recon_batch, X_train_tensor, mean, log_var)
        loss.backward()
        optimizer.step()

    model.eval()
    with torch.no_grad():
        recon_test, _, _ = model(X_test_tensor)
        reconstruction_error = torch.mean((recon_test - X_test_tensor) ** 2, dim=1).numpy()

    scores = (reconstruction_error - np.min(reconstruction_error)) / (np.max(reconstruction_error) - np.min(reconstruction_error))

    fpr, tpr, thresholds = roc_curve(y_true, scores)
    auc_score = auc(fpr, tpr)
    ap_score = average_precision_score(y_true, scores)

    auc_scores.append(auc_score)
    ap_scores.append(ap_score)

mean_auc = np.mean(auc_scores)
mean_ap = np.mean(ap_scores)

results_SPF.loc["SPF", "VAE_AUC"] = mean_auc
results_SPF.loc["SPF", "VAE_AP"] = mean_ap

```

```python
faults = ['Pastry', 'Z_Scratch', 'K_Scratch', 'Stains', 'Dirtiness', 'Bumps', 'Other_Faults']

results_SPF = pd.DataFrame(index=["SPF"])

apcDl = []
apcDFl=[]
aucDl = []
aucDFl = []

# VAE Model Definition
class VAE(nn.Module):
    def __init__(self, input_dim, latent_dim=10):
        super(VAE, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU()
        )
        self.z_mean = nn.Linear(32, latent_dim)
        self.z_log_var = nn.Linear(32, latent_dim)
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 32),
            nn.ReLU(),
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, input_dim),
            nn.Sigmoid()
        )

    def encode(self, x):
        h = self.encoder(x)
        return self.z_mean(h), self.z_log_var(h)

    def reparameterize(self, mean, log_var):
        std = torch.exp(0.5 * log_var)
        eps = torch.randn_like(std)
        return mean + eps * std

    def decode(self, z):
        return self.decoder(z)

    def forward(self, x):
        mean, log_var = self.encode(x)
        z = self.reparameterize(mean, log_var)
        return self.decode(z), mean, log_var

def loss_function(recon_x, x, mean, log_var):
    recon_loss = nn.MSELoss()(recon_x, x)
    kld_loss = -0.5 * torch.sum(1 + log_var - mean.pow(2) - log_var.exp())
    return recon_loss + kld_loss

for fault in faults:
    data = steel_plates_faults.data.original


    non_faults = [f for f in faults if f != fault]

    Xn = data[data[fault] == 1].drop(columns=faults)

    Xa = data[data[non_faults].sum(axis=1) == 0].drop(columns=faults)

    # Normalize the data
    scaler = StandardScaler()
    Xn_scaled = scaler.fit_transform(Xn)
    Xa_scaled = scaler.transform(Xa)
    # Convert to tensors
    Xn_tensor = torch.FloatTensor(Xn_scaled)
    Xa_tensor = torch.FloatTensor(Xa_scaled)
    # Define and train the VAE
    input_dim = Xn_scaled.shape[1]
    latent_dim = 10
    vae_model = VAE(input_dim=input_dim, latent_dim=latent_dim)
    optimizer = optim.Adam(vae_model.parameters(), lr=0.001)

    print("Training VAE...")
    vae_model.train()
    for epoch in range(50):  # Adjust epochs as needed
        optimizer.zero_grad()
        recon_batch, mean, log_var = vae_model(Xn_tensor)
        loss = loss_function(recon_batch, Xn_tensor, mean, log_var)
        loss.backward()
        optimizer.step()
        if epoch % 10 == 0:
            print(f"Dataset {i} | Epoch {epoch} | Loss: {loss.item()}")

    # Extract features using the encoder
    print(f"Encoding dataset {i}...")
    vae_model.eval()
    with torch.no_grad():
        Xn_encoded = vae_model.encoder(Xn_tensor).numpy()
        Xa_encoded = vae_model.encoder(Xa_tensor).numpy()

    print(f"Dataset {i} encoded.")

    # Train Diff-RF on encoded features
    print(f"Training Diff-RF on dataset {i}...")
    X_train_enc, X_test_enc= train_test_split(Xn_encoded, test_size=0.2, random_state=42)

    X_test = np.vstack((X_test_enc, Xa_encoded))
    y_true = np.array([-1] * len(X_test_enc) + [1] * len(Xa_encoded))

    size = int(0.25*len(X_train_enc))
    ntrees= 128
    model = DiFF_TreeEnsemble(sample_size=size, n_trees=ntrees)    # load data


    X_train_enc = np.array(X_train_enc,dtype=float)
    Xa_enc = np.array(Xa_encoded,dtype=float)


    # Fit the model and compute scores
    model.fit(X_train_enc,n_jobs=8)

# Create y_true labels

    sc_di, sc_ff, sc_diff_rf = model.anomaly_score(X_test, alpha_value)
    sc_diff_rf=np.array(sc_diff_rf)
    sc_di=np.array(sc_di)

    sc_di=(sc_di-sc_di.min())/(sc_di.max()-sc_di.min())
    sc_diff_rf=(sc_diff_rf-sc_diff_rf.min())/(sc_diff_rf.max()-sc_diff_rf.min())



    fpr_D, tpr_D, thresholds = roc_curve(y_true, sc_di)
    aucD=auc(fpr_D, tpr_D)
    fpr_DF, tpr_DF, thresholds = roc_curve(y_true, sc_diff_rf)
    aucDF=auc(fpr_DF, tpr_DF)
    apcD = average_precision_score(y_true, sc_di)
    apcDF = average_precision_score(y_true, sc_diff_rf)

    apcDl.append(apcD)
    apcDFl.append(apcDF)
    aucDl.append(aucD)
    aucDFl.append(aucDF)

mean_auc_pw = np.mean(aucDl)
mean_auc_cl = np.mean(aucDFl)
mean_ap_pw = np.mean(apcDl)
mean_ap_cl = np.mean(apcDFl)


results_SPF.loc["SPF", "Point-wise AE_Diff-RF_AUC"] = mean_auc_pw
results_SPF.loc["SPF", "Point-wise AE_Diff-RF_AP"] = mean_ap_pw


results_SPF.loc["SPF", "Collective AE_Diff-RF_AUC"] = mean_auc_cl
results_SPF.loc["SPF", "Collective AE_Diff-RF_AP"] = mean_ap_cl
```
